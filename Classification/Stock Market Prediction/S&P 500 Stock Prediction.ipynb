```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history("max")
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y = "Close", use_index = True)
```

```python
sp500.drop(["Dividends","Stock Splits"], axis= 1,inplace= True)
```

```python
sp500.head()
```

- Predict stock price up or down tomorrow

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"]> sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":]
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators= 100, min_samples_leaf=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:] 

predictors = ["Close","Volume","Open","High","Low"]
model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
import pandas as pd

preds = pd.Series(preds,index= test.index)
preds
```

```python
precision_score(preds,test["Target"])
```

```python
combined = pd.concat([preds,test["Target"]], axis = 1)
```

```python
combined.plot()
```
