```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# vẽ biểu đồ mô tả số ca nhiễm, số ca hồi phục, số ca tử vong

## Cách 1

```python
with open('covid_19.csv','r') as f:
    file = f.readlines()
```

```python
total_matrix = []
for it in file:
    val = it.rstrip('\n')
    val = val.split(',')
    total_matrix.append(val)
    
total_matrix
```

```python
lst_confirmed = []
lst_deaths = []
lst_recoveries = []
lng = len(total_matrix)
lng
```

```python
for i in range(1,lng):
        lst_confirmed.append(total_matrix[i][0])
        lst_deaths.append(total_matrix[i][1])
        lst_recoveries.append(total_matrix[i][2])
```

```python
lst_confirmed = [int(i) for i in lst_confirmed]
lst_deaths = [int(i) for i in lst_deaths]
lst_recoveries = [int(i) for i in lst_recoveries]
```

```python
lst_confirmed = np.array(lst_confirmed)
lst_deaths = np.array(lst_deaths)
lst_recoveries = np.array(lst_recoveries)
```

```python
plt.plot(lst_confirmed)
plt.xlabel('number of day')
plt.ylabel('# of confirmed cases')
```

```python
plt.plot(lst_deaths)
plt.xlabel('number of day')
plt.ylabel('# of confirmed cases')
```

```python
plt.plot(lst_recoveries)
plt.xlabel('number of day')
plt.ylabel('# of confirmed cases')
```
