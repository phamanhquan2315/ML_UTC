```python
def fluctuate(total):
    ans = np.array(0)
    for i in range(1,len(total)):
        ans =np.append(ans,total[i] - total[i-1])
    return ans
```

```python
# DataFrame thể hiện sự chênh lệch giữa phút sau và phút trước với giá cao nhất
df1 = df1.apply(fluctuate,axis = 0)
```
